```python
for j in range(59146):
    if(j%1000 == 0):
        print(j)
    article = str(data["Text"][j]).lower()
    tokens = word_tokenize(article)
    article_score = 0
    for word in tokens:
        for i in range(31):
            if( word == associated_words[i]):
                article_score += associated_weights[i]
    data.at[j, 'Score'] = article_score
```
