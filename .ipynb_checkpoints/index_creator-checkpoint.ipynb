```python
from nltk.tokenize import word_tokenize
```

```python
# This iterates through each word in the tokenized article and scores it

article = data["Text"][160].lower()

tokens = word_tokenize(article)
article_score = 0
for word in tokens:
    for i in range(31):
        if( word == associated_words[i]):
            print(word)
            article_score += associated_weights[i]
        
print(article_score)
```

```python
data["Score"] = data.apply(lambda row: 0) 
```

```python
data.head(10)
```

```python
# Running the scoring for all articles in an year

length = len(data)
for j in range(length):
    if(j%1000 == 0):
        print(j)
    article = str(data["Text"][j]).lower()
    tokens = word_tokenize(article)
    article_score = 0
    for word in tokens:
        for i in range(31):
            if( word == associated_words[i]):
                article_score += associated_weights[i]
    data.at[j, 'Score'] = article_score
```

```python
# Sorting based on scores to see the highest score we have
data = data.sort_values(by="Score", ascending=False).reset_index(drop=True)
```

```python
# The value 1.5 was chosen as the any article would need atleast two words to get past this score threshold
num_corr1 = data[data["Score"]>1.5]
print(len(num_corr1))
```

```python
num_corr2 = data[data["Score"]>1.75]
print(len(num_corr2))
```

```python
# Calculating the percentage of articles in a year which cross the threshold
fract1 = len(num_corr1)/len(data)
print(fract1)
fract2 = len(num_corr2)/len(data)
print(fract2)
```

```python
#Saving fractions of each year to a txt file for further analysis

filehandle= open("Yearwise.txt",'a')
filehandle.write("\n\nYear: 2009\n")
filehandle.write("Fract1: "+str(fract1))
filehandle.write("\nFract2: "+str(fract2))
filehandle.close()
```
