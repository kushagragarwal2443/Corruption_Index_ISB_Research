```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./y17_States_set_with_spaces.csv").drop("Unnamed: 0", axis=1).reset_index(drop=True)
data.head(10)
```

```python
data["State_final"]=data.apply(lambda row: None)
```

```python
data.head(100)
```

```python
# I have used the state information given by the hindu to wherever available as the final state tagging for the article
# If the hindu hasnt tagged the article then i used my state_tagging.ipynb's results to tag a state to the article
for i in range(len(data)):
    
    # if both hindu tag and my_state tag are same then final_state = hindu_tag
    if(data["State"][i].lower() == data["Category"][i]):
        data["State_final"][i]= data["State"][i].lower()
        
    # hindu labels articles as hyderabad, mumbai and bengaluru as well, hence manually considering for those cases
    elif(data["State"][i].lower() == "andhra pradesh" and data["Category"][i] == "hyderabad"):
        data["State_final"][i]= data["State"][i].lower()
    elif(data["State"][i].lower() == "maharashtra" and data["Category"][i] == "mumbai"):
        data["State_final"][i]= data["State"][i].lower()
    elif(data["State"][i].lower() == "karnataka" and data["Category"][i] == "bengaluru"):
        data["State_final"][i]= data["State"][i].lower()
        
    # if hindu hasnt labelled the data then final_state = my_state
    elif(data["Category"][i]=='national' or  data["Category"][i]=='economy' or data["Category"][i]=='business' or data["Category"][i]=='states' or data["Category"][i]=='other states'):
        data["State_final"][i] = data["State"][i].lower()
        
    # again controlling for bengaluru hyderabad and mumbai in the hindu tags
    elif(data["Category"][i]=="bengaluru"):
        data["State_final"][i] = "karnataka"
    elif(data["Category"][i]=="hyderabad"):
        data["State_final"][i] = "andhra pradesh"
    elif(data["Category"][i]=="mumbai"):
        data["State_final"][i] = "maharashtra"
        
    # final case, if hindu_tag and my_state are different, then final_state = hindu_tag
    elif(data["State"][i].lower() != data["Category"][i]):
        data["State_final"][i] = data["Category"][i]
        
data.head(100)
```

```python
state_groups = data.groupby("State_final")
state_list = data["State_final"].unique()
print(len(state_list))
for states in state_list:
    print(states, len(state_groups.get_group(states)))
```

```python
# cutoff is sum of two least weighted words
cutoff = 0.566376268863678+0.5651700496673584
print(cutoff)
```

```python
# The csvs have the states in this order
statelist_fromcsv=['andhra pradesh', 'arunachal pradesh', 'assam', 'bihar', 'jharkhand', 'goa', 'gujarat', 'haryana',
'himachal pradesh', 'jammu & kashmir', 'karnataka', 'kerala', 'madhya pradesh', 'chhattisgarh', 'maharashtra',
'manipur', 'meghalaya', 'mizoram', 'nagaland', 'orissa', 'punjab', 'rajasthan', 'sikkim', 'tamil nadu', 'telangana',        
'tripura', 'uttar pradesh', 'uttaranchal', 'west bengal', 'andaman & nicobar islands', 'chandigarh', 'delhi', 'pondicherry']
```

```python
dictionary= dict(zip(statelist_fromcsv, np.zeros(len(statelist_fromcsv))))
print(len(dictionary))
```

```python
# create statewise ratios of articles crossing the threshold
count=0
for states in state_list:
    state_data = state_groups.get_group(states)
    x = len(state_data[state_data["Score"] > cutoff])
    y= len(state_data)
    
    # take states which have 100 articles from it
    if(states in statelist_fromcsv and y>100):
        count+=1
        dictionary[states]=x/y
print(count)
```

```python
# Simply copy from the csv file as it is. ALl modifications have been made in the code above
state_gdp_growth=[8.8,9.68, 12.24, 15.01, 9.66, 9.41, 13.09, 13.02, 8.94, 12.34, 11.34, 13.7, 11.41, 11.13, -1, 10.13, 
                  11.69, 15.07, 11.4, 14.05, 10.13, 12.86, 14, 13.84, 14.33, 13.97, 14.23, 10.35, 17.81, -1, 11.52, 12.98, 
                  11.21]
```

```python
for i in range(len(state_gdp_growth)):
    print(statelist_fromcsv[i], state_gdp_growth[i])
```

```python
updated_state_list=[]
corr_index=[]
sgdp=[]
for i in range(len(state_gdp_growth)):
    statei = statelist_fromcsv[i]
    sgdpi = state_gdp_growth[i]
    corri=dictionary[statei]
    
    # if both corruption information and state_gdps are available
    if(corri>0 and sgdpi>0):
        updated_state_list.append(statei[:5])
        
        # scaling the corruption inverse 
        corr_index.append(0.21-corri)
        
        # convert sgdp from percentage to ratios
        sgdp.append(sgdpi/100)
```

```python
for i in range(len(sgdp)):
    print(updated_state_list[i], corr_index[i], sgdp[i])
```

```python
# Plot corruption index inverse against sgdp shifted for one year of impact
plt.figure(figsize=(14,5))
plt.plot(updated_state_list, sgdp, label="SGDP")
plt.plot(updated_state_list, corr_index, label="Corr")
plt.scatter(updated_state_list, sgdp)
plt.scatter(updated_state_list, corr_index)
plt.legend(loc='best')
plt.show()
```
