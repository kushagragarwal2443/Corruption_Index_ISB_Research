```python
import pandas as pd
import numpy as np
```

```python
final_dict = dict()
```

```python
cutoff = 0.566376268863678+0.5651700496673584
statelist_fromcsv=['andhra pradesh', 'arunachal pradesh', 'assam', 'bihar', 'jharkhand', 'goa', 'gujarat', 'haryana',
'himachal pradesh', 'jammu & kashmir', 'karnataka', 'kerala', 'madhya pradesh', 'chhattisgarh', 'maharashtra',
'manipur', 'meghalaya', 'mizoram', 'nagaland', 'orissa', 'punjab', 'rajasthan', 'sikkim', 'tamil nadu', 'telangana',        
'tripura', 'uttar pradesh', 'uttaranchal', 'west bengal', 'andaman & nicobar islands', 'chandigarh', 'delhi', 'pondicherry']
```

```python
for year in range(12,18):
    filename = "./Complete_Data/y"+str(year)+"_States_set_with_spaces.csv"
    data = pd.read_csv(filename)
    print(year, len(data))
    data["State_final"]=data.apply(lambda row: None)
    for i in range(len(data)):
        if(data["State"][i].lower() == data["Category"][i]):
            data["State_final"][i]= data["State"][i].lower()
        elif(data["State"][i].lower() == "andhra pradesh" and data["Category"][i] == "hyderabad"):
            data["State_final"][i]= data["State"][i].lower()
        elif(data["State"][i].lower() == "maharashtra" and data["Category"][i] == "mumbai"):
            data["State_final"][i]= data["State"][i].lower()
        elif(data["State"][i].lower() == "karnataka" and data["Category"][i] == "bengaluru"):
            data["State_final"][i]= data["State"][i].lower()
        elif(data["Category"][i]=='national' or  data["Category"][i]=='economy' or data["Category"][i]=='business' or data["Category"][i]=='states' or data["Category"][i]=='other states'):
            data["State_final"][i] = data["State"][i].lower()
        elif(data["Category"][i]=="bengaluru"):
            data["State_final"][i] = "karnataka"
        elif(data["Category"][i]=="hyderabad"):
            data["State_final"][i] = "andhra pradesh"
        elif(data["Category"][i]=="mumbai"):
            data["State_final"][i] = "maharashtra"
        elif(data["State"][i].lower() != data["Category"][i]):
            data["State_final"][i] = data["Category"][i]
    state_groups = data.groupby("State_final")
    state_list = data["State_final"].unique()
    dictionary= dict(zip(statelist_fromcsv, np.zeros(len(statelist_fromcsv), dtype=int)))
    count=0
    for states in state_list:
        state_data = state_groups.get_group(states)
        x = len(state_data[state_data["Score"] > cutoff])
        y= len(state_data)
        if(states in statelist_fromcsv and y>100):
            count+=1
            dictionary[states]=x/y
    print(dictionary,"\n\n")
    final_dict[year] = dictionary
    
    
```
